```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
```

```python
mpl.rc('font', family='Malgun Gothic')
plt.rc('font', family='Malgun Gothic')
```

```python
DIDB_STATS_PATH = np.load("C:/Users/rlfalsgh95/source/repos/PRNU/PRNU_DIDB/digital_camera_image/stats.npy", allow_pickle = True).item()
SMDB_DIGITAL_STATS_PATH = np.load("C:/Users/rlfalsgh95/source/repos/PRNU/PRNU_SMDB/digital_camera_image/stats.npy", allow_pickle = True).item()
SMDB_SMARTPHONE_STATS_PATH = np.load("C:/Users/rlfalsgh95/source/repos/PRNU/PRNU_SMDB/smartphone_image/stats.npy", allow_pickle = True).item()
```

```python
matplotlib.style.use('classic')
matplotlib.style.use('seaborn-white')
```

```python
dpi = 1000
plt.figure(figsize=(10,10), dpi= dpi)
plt.grid(True, color = "black")

FONTSIZE = 25

line = plt.plot(SMDB_DIGITAL_STATS_PATH["fpr"][0::10], SMDB_DIGITAL_STATS_PATH["tpr"][0::10], "-", color = "red", label = "SMDB Digital camera")
plt.setp(line, linewidth=3.0)
line = plt.plot(DIDB_STATS_PATH["fpr"][0::200], DIDB_STATS_PATH["tpr"][0::200], "--", color = "tab:purple", label = "DIDB Digital camera")
plt.setp(line, linewidth=4.0)
line = plt.plot(SMDB_SMARTPHONE_STATS_PATH["fpr"][0::90], SMDB_SMARTPHONE_STATS_PATH["tpr"][0::90], "o-", color = "tab:cyan", label = "SMDB Smartphone")
plt.setp(line, linewidth=3.0)

plt.legend(loc = "lower right", frameon=True, fontsize = FONTSIZE)
ticks = np.arange(0, 1.1, 0.1)
plt.xticks(np.arange(0, 1.1, 0.1), fontsize = FONTSIZE)
plt.yticks(np.arange(0.1, 1.1, 0.1), fontsize = FONTSIZE)
plt.xlabel("False Positive Rate", fontsize= FONTSIZE)
plt.ylabel("True Positive Rate", fontsize= FONTSIZE)
plt.savefig("C:/Users/rlfalsgh95/source/repos/PRNU/PRNU.png")
```
